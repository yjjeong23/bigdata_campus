```python
uniqNouns = []

for file in fileList:
    with open(file, "r", encoding='utf-8') as fp:
        content = fp.read().strip()
        content = re.sub(r"[\s]{2:}","", content)
    
    for sentence in nltk.sent_tokenize(content):
        nouns=[]
        for word in nltk.word_tokenize(sentence):
            noun = Kkma().nouns(word)
            nouns.extend(noun)
            nouns = list(set(nouns))

        uniqNouns.extend(nouns)
        uniqNouns = list(set(uniqNouns))
#         print(sentence)
    #     print(word)
#         print(nouns)
    #     words = Kkma().morphs(sentence)
    #     print(words)
print(len(uniqNouns))
```

```python
docNouns = {}
maxCount = {}


for file in fileList:
    docid = file[-17:]
    with open(file, "r", encoding="utf-8") as fp:
        maxCount[docid] = 0
        content = fp.read().strip()
        content = re.sub(r"[\s]{2,}","", content)
        
    nouns={}
    
    for sentence in nltk.sent_tokenize(content):
        for word in nltk.word_tokenize(sentence):
            for w in Kkma().nouns(word):
                if w in nouns.keys():
                    nouns[w] +=1
                else:
                    nouns[w] = 1
                    
                if maxCount[docid] < nouns[w]:
                    maxCount[docid] = nouns[w]
                    
    docNouns[docid] = nouns
#         uniqNouns.extend(nouns)
#         uniqNouns = list(set(nouns))
#         print(sentence)
#     #     print(word)
#         print(nouns)
#     #     words = Kkma().morphs(sentence)
#     #     print(words)
# print(len(uniqNouns))
```

```python
for k,v in docNouns.items():
    print("{0} - {1}".format(k, len(v)))
```

```python
invertedIndex = {}

for noun in uniqNouns:
    invertedIndex[noun] = []
    
    for k,v in docNouns.items():
        if noun in v:
            invertedIndex[noun].append(k)
```

```python
#가중치 계산을 통한 feature 추출

kRation = 0.5
TF = {}

for k, v in docNouns.items():
    tfList = {}
    
    for w in v:
        tfList[w] = kRation + (1-kRation) * (v[w] / maxCount[k])
        print("{0} | {1} + {2} + ({3} / {4}) = {5}".format(w, kRation, (1-kRation), v[w], maxCount[k], tfList[w]))
#         v[w]   #해당 문서에서 각 단어의 빈도
    
    TF[k] = tfList
```
