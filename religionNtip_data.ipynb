stack / unstack > 데이터의 구조를 바꿔줌
- stack : 원래의 컬럼들을 row로 변경해서 세로로 긴 형태의 데이터로 변경 - series를 반환함
- unstack: row에 있는 변수들을 column으로 변경해서 가로로 긴 형태의 데이터로 변경

```python
df= df.stack()    #일렬로 만들어줌 - chart 만들 때 유용 > series를 반환
```

- pivot table : 인자로 aggregation함수를 받아 적용 후 변경된 값이 들어가 있는 테이블로 변경
- pivot : 새로운 column으로 설정한 컬럼의 값들이 column명으로 들어감 (구조를 변경하는 것은 맞음) - 단, 하나의 entry에 복수의 값이 들어가야 하는 경우 에러 발생 주의


```python
data.pivot_table(index=['smoker','sex'], values='tip')
```

```python
graph = pd.pivot_table(data, values='tip', index='sex')
```

```python
graph.plot(kind = 'barh')
```

```python
graph.plot.pie(subplots=True)
```
