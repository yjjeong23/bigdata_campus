```python
import pandas as pd
```

```python
rate_col = ['user_id','movie_id','rating','unix_timestamp']
ratings = pd.read_csv("ml-100k/u.data", sep="\t", names=rate_col)
```

```python
ratings['date'] = pd.to_datetime(ratings['unix_timestamp'], unit='s')
```

```python
ratings.head()
```

```python
item_col = ['movie_id','title','release_date','video_release_date', 'lmdb_url']
item = pd.read_csv("ml-100k/u.item" , sep="|",names=item_col , encoding='Latin1', usecols=range(5))
```

```python
item.head()
```

```python
user_col = ['user_id','age','sex','occupation','zip_code']
user = pd.read_csv("ml-100k/u.user", sep="|", names=user_col)
```

```python
user.head()
```

```python
ratings.info()
```

```python
ratings.describe()
```

```python
movie_rating = pd.merge(item, ratings)
lens = pd.merge(movie_rating, user)
lens.head()
```

```python
lens.sort_values('rating', ascending=False).head()
```

```python
#rating이 가장 많은 영화 TOP 10
lens.title.value_counts()[:10]
```

```python
lens.title.value_counts().nlargest(10)
```

```python
lens.groupby('title').mean().rating.head(30)
```

```python
lens.groupby('title').size().sort_values(ascending=False)
```

```python
lens.groupby('user_id').size().sort_values(ascending=False)
```

```python
lens.groupby(['user_id','movie_id']).mean()
```

```python
import numpy as np
```

```python
movie_stats = lens.groupby('title').agg({'rating': [np.size, np.mean]})   #여러 통계를 한번에 낼 때 agg 함수 사용
movie_stats.sort_values([('rating','mean')], ascending=False).head()
```

```python
#조건 마스킹
atleast_100 = movie_stats['rating']['size'] >=100
movie_stats[atleast_100].sort_values([('rating','mean')], ascending=False).head()
```
