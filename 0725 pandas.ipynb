```python
#attribute 방식
df.title
df.columns

#dictionary 방식 - 두개 이상도 한 번에 가져오려면 인자를 list로 묶어줌
df[['title','rating']]

#index로 접근은 불가
#df[0] - (X)

#n번째 row 출력(인덱스방식+slicing 지원) - 두개 이상 한 번에 가져오려면 인자를 list로 묶어줌
df.loc[0]
df.loc[[0,3]]
df.loc[1:3, ['title', 'rating']]
#column은 slicing이 안되지만 row는 가능
df.loc[:,['title','rating']]  #하면 dictionary 방식과 같음. col에 해당하는 모든 row 출력

#df.loc는 label base (이름을 기준으로 slicing)
#df.at 는 label base이지만 한 개만 가져올 때 사용 - indexing

%timeit df.loc[1,'rating']  #iteration하면서 time check - 성능체크. 
%timeit df.at[1,'rating'] # loc보다 at이 속도가 빠름

df.loc[1, 'title'] ='abc' #하면 해당 value가 바뀜. col 명도 바꿀 수 있지만 권장하지 않음

df.axes
df.iat[1,1]  #정수값 인덱스로 딱 한 개 가져옴. dictionary방식을 index로 변환하는 과정을 필요하기때문에 df.at보다 약간 느림
df.iloc[1:3, 1]  #slicing할 때 파이썬 list slicing 처럼 맨 마지막 값은 빼고 출력
```
