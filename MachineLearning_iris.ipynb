```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.iloc[:,:-1], iris.iloc[:,-1], test_size=0.3)  #randome_state를 지정하지 않으면 랜덤하게 split
```

```python
len(X_train)
```

```python
knn.fit(X_train, y_train)
```

```python
knn.predict(X_test)
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, knn.predict(X_test))  # 행렬의 대각선값이 정확하게 예측한 갯수
```

cross-validation을 통해 전체 데이터셋을 여러 방법으로 split -> overfitting을 막을 수 있는 parameter 를 찾아냄

```python
knn.score(iris.iloc[:,:-1], iris.iloc[:,-1])  #성능 측정
```
